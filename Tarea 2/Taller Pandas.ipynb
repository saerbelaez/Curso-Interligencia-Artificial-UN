#Manipulación datos textuales (Limpieza)

Para realizarle limpieza a la base de datos, se puede usar lo siguiente:



*   *lower*: Poner en minúscula todo el texto.
*   *replace*: Reemplazar ciertos valores por otros.
*   *strip*: Eliminar los espacios al principio y al final de la cadena.
*   *title*: Poner primera letra de cada palabra en mayúscula.



````python
Matricula["INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)"] = (Matricula["INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)"]
                                                  .str.lower()
                                                  .str.replace("[^a-záéíóúüñ ]","")
                                                  .str.replace(" +"," ")
                                                  .str.strip()
                                                  .str.title()
                                                  ) 
````

````python
print(Matricula)
````

#**Modificar tablas**

Es posible hacer cambios de orden y estructura en las tablas de pandas

Organizar de forma descendiente en función de la cantidad de matriculados

````python
Matricula.sort_values(["MATRICULADOS"],ascending=False)
````

Usando *groupby()* podemos juntar los datos por año y sumarlos para tener el total de matriculados por IES. 

````python
Matricula.groupby("IES_PADRE")[["MATRICULADOS"]].sum()
````

Podemos aplicar ambos métodos en un mismo código.

````python
Matricula_sort = (Matricula.groupby("IES_PADRE")[["MATRICULADOS"]]
                  .sum()
                  .sort_values("MATRICULADOS",ascending = False))
display(Matricula_sort)
````

#**Concatenar y unir**

Es posible unir tablas tanto vertical como horizontalmente.

En ambos casos, podemos usar *concat()*, y se juntarán las bases con base en los nombres de columnas o los índices de las filas. 

````python
Matricula_sem1=Matricula[Matricula["SEMESTRE"]==1]
display(Matricula_sem1)
````

````python
Matricula_sem2=Matricula[Matricula["SEMESTRE"]==2]
Matricula_sem2["dummy"]=0
display(Matricula_sem2)
````

````python
Mat_sem1y2=pd.concat([Matricula_sem1,Matricula_sem2])
display(Mat_sem1y2)
````

Si se usa *merge()*, podemos hacer uniones de las tablas de forma horizontal que compartan una columna/índice en común.

El comando puede especificar el tipo de cruce que se desee realizar con *how*, las opciones son *outer*,  *left* y *right*

````python
outer_joined=pd.merge(Matricula_sem1, Matricula_sem1, how='outer')
outer_joined.head(10)
````

#**Series de tiempo y time stamps**

Existe un tipo de datos llamado "timestamp" que se usa para medir registros de tiempo. Es posible pasar strings a este tipo de dato en diferentes formatos y trabajar con este usándolo como indice

````python
ice=pd.read_csv("gdrive/My Drive/Maestría/Tesis/Datos finales/Indice de confianza empresarial.csv",delimiter=",")
ice.head()
````

````python
ice["TIME"]=pd.to_datetime(ice["TIME"],format="%Y-%m")
ice.index=ice["TIME"]
ice=ice.drop(["TIME"],axis=1)
ice.head
````

Al definir la fecha como índice, se puede extraer información con respecto al año y al mes.

````python
ice.groupby([ice.index.month])["Value"].describe()
````

````python
ice.groupby([ice.index.year]).sum("Value").head(10)
````
